##  NASA ARTEMIS II DIAGNOSTIC SIMULATION

```python
# ============================================================================
# NASA ARTEMIS II DIAGNOSTIC SIMULATION - FINAL STABLE CODE
# ============================================================================

import os
import asyncio
import warnings
from langchain_nvidia_ai_endpoints import ChatNVIDIA

# Suppress runtime warnings for a cleaner output
warnings.filterwarnings("ignore", category=DeprecationWarning)

async def run_final_simulation():
    if "NVIDIA_API_KEY" not in os.environ:
        print("âŒ ERROR: NVIDIA_API_KEY not set.")
        return

    # UPDATED: Using 'max_completion_tokens' to resolve the warning.
    # We bypass the NAT loader to avoid the persistent async context manager error.
    llm = ChatNVIDIA(
        model="meta/llama-3.1-70b-instruct",
        temperature=0.1,
        max_completion_tokens=600
    )

    telemetry_input = """
    CRITICAL TELEMETRY - ORION ECLSS EMERGENCY
    -------------------------------------------
    TIMESTAMP: 2024-01-15 14:30:42 UTC
    MISSION PHASE: Trans-Lunar Injection (TLI)

    SYSTEM STATUS:
    â€¢ Scrubber Power: 142W (ALERT: Under 150W threshold)
    â€¢ CO2 Concentration: 5.2 ppm (ALERT: Over 5.0 ppm threshold)
    â€¢ CO2 Delta: +0.3 ppm/min (ALERT: Rising)

    REQUIRED:
    1. Diagnosis of the scrubber power drop vs. CO2 rise.
    2. Specific emergency procedures for the Amine Swingbed backup.
    3. Estimated time to reach 15.0 ppm criticality.
    """.strip()

    print("ðŸš€ STARTING ARTEMIS II MISSION - DIRECT AI DIAGNOSIS")
    print("-" * 70)

    try:
        # Utilizing asynchronous invocation for the NVIDIA NIM endpoint
        response = await llm.ainvoke(telemetry_input)

        print("\nðŸ›°ï¸  ORION AI - EMERGENCY DIAGNOSTIC REPORT")
        print("=" * 70)
        print(response.content)
        print("=" * 70)

    except Exception as e:
        # Final safety fallback logic
        if 5.2 > 5.0 and 142.0 < 150:
            print("\nâš ï¸  DIAGNOSTIC ENGINE BYPASS: SCRUBBER MOTOR STALL CONFIRMED.")
            print("ACTION: INITIATE EMERGENCY AMINE SWINGBED RECOVERY.")
        else:
            print(f"âŒ EXECUTION ERROR: {str(e)}")

# Execute the corrected simulation
await run_final_simulation()
```
