## H2E

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as sci
from scipy.spatial.distance import pdist

# Using Seed 123 for exact reproducibility in H2E validation
np.random.seed(123)

# ---------------------------------------------------------
# 1. DISTANCE CONCENTRATION (The "Metric Collapse" Problem)
# ---------------------------------------------------------
def distance_ratio(x):
    """
    Calculates the ratio of the largest to smallest Euclidean distance.
    In H2E, this proves why standard loss functions fail in high-D,
    justifying the move toward Riemannian geometry.
    """
    # Calculate all pairwise Euclidean distances
    distances = pdist(x.T, metric='euclidean')

    smallest_dist = np.min(distances)
    largest_dist = np.max(distances)

    dist_ratio = largest_dist / smallest_dist
    return dist_ratio

def plot_distance_distributions():
    dims = [2, 10, 100, 1000]
    plt.figure(figsize=(10, 5))

    for d in dims:
        x = np.random.normal(size=(d, 500))
        distances = pdist(x.T)
        # Normalize by sqrt(d) to compare distributions on the same scale
        plt.hist(distances / np.sqrt(d), bins=50, alpha=0.5, label=f'{d}D')

    plt.title("Distance Concentration (Seed 123): The Euclidean Breakdown")
    plt.xlabel("Normalized Euclidean Distance")
    plt.ylabel("Frequency")
    plt.legend()
    plt.grid(alpha=0.3)
    plt.show()

# ---------------------------------------------------------
# 2. HYPERSPHERE VOLUME (The "Vanishing Expert Space")
# ---------------------------------------------------------
def volume_of_hypersphere(diameter, dimensions):
    """
    Calculates volume using the Gamma function.
    Proves that "expert reasoning" occupies near-zero volume relative
    to the total embedding space, supporting model compression.
    """
    radius = diameter / 2.0
    pi = np.pi

    numerator = np.power(pi, dimensions / 2.0)
    denominator = sci.gamma(dimensions / 2.0 + 1)

    volume = (numerator / denominator) * np.power(radius, dimensions)
    return volume

def plot_volume_decay():
    dims = np.arange(1, 21)
    volumes = [volume_of_hypersphere(1.0, d) for d in dims]

    plt.figure(figsize=(10, 5))
    plt.plot(dims, volumes, 'o-', color='teal', linewidth=2, label='Hypersphere Volume')
    plt.fill_between(dims, volumes, color='teal', alpha=0.1)
    plt.axhline(y=1.0, color='r', linestyle='--', label='Unit Cube Volume (1.0)')

    plt.title("The Vanishing Volume: Why H2E Requires Precision")
    plt.xlabel("Dimensions")
    plt.ylabel("Volume")
    plt.xticks(dims)
    plt.legend()
    plt.grid(True, linestyle=':', alpha=0.6)
    plt.show()

# ---------------------------------------------------------
# 3. THE H2E SHELL EFFECT (Topological Certainty)
# ---------------------------------------------------------
def get_prop_of_volume_in_outer_1_percent(dimension):
    """
    Core of Topological Certainty: Proves that in high dimensions,
    data (and expert knowledge) lives on the surface/shell.
    """
    proportion = 1.0 - np.power(0.99, dimension)
    return proportion

def plot_shell_concentration():
    dims = np.linspace(1, 500, 100)
    proportions = [get_prop_of_volume_in_outer_1_percent(d) for d in dims]

    plt.figure(figsize=(10, 5))
    plt.plot(dims, proportions, color='orangered', linewidth=2.5)
    plt.axhline(y=0.951, color='navy', linestyle=':', label='95.1% Threshold (300D)')

    plt.title("H2E Shell Effect: Concentration on the Manifold Boundary")
    plt.xlabel("Dimensions")
    plt.ylabel("Proportion of Total Volume")
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()

# ---------------------------------------------------------
# EXECUTION AND OUTPUT
# ---------------------------------------------------------

# Part 1: Dataset Generation and Ratios
n_data = 1000
x_2D = np.random.normal(size=(2, n_data))
x_100D = np.random.normal(size=(100, n_data))
x_1000D = np.random.normal(size=(1000, n_data))

print(f"--- Statistics for Seed 123 ---")
print(f'Ratio 2D: {distance_ratio(x_2D):.3f}')
print(f'Ratio 100D: {distance_ratio(x_100D):.3f}')
print(f'Ratio 1000D: {distance_ratio(x_1000D):.3f}')

# Part 2: Hypersphere Volume Decay
print("\n--- Hypersphere Volume Decay ---")
diameter = 1.0
for c_dim in range(1, 11):
    vol = volume_of_hypersphere(diameter, c_dim)
    print(f"Volume in {c_dim}D: {vol:.3f}")

# Part 3: Shell Effect
print("\n--- The H2E Shell Effect ---")
for c_dim in [1, 2, 10, 20, 50, 100, 150, 200, 250, 300]:
    prop = get_prop_of_volume_in_outer_1_percent(c_dim)
    print(f'Outer 1% shell volume in {c_dim}D = {prop:.3f}')

# Part 4: Visual Proofs
plot_distance_distributions()
plot_volume_decay()
plot_shell_concentration()
```
