```python
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse, Arc, FancyArrowPatch, Circle, Rectangle
import numpy as np
import matplotlib.patheffects as path_effects

plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(12, 10), facecolor='#0a0a1f')
ax.set_facecolor('#0a0a1f')

# Manifold ellipses
center = (0.5, 0.55)
ax.add_patch(Ellipse(center, 1.05, 0.75, color='#111827', alpha=0.88, ec='#3b82f6', lw=3.5))
ax.add_patch(Ellipse(center, 0.85, 0.6, color='#1e293b', alpha=0.93, ec='#60a5fa', lw=2.8))

# Hexagonal lattice
angles = np.linspace(0, 360, 7)[:-1]
for ang in angles:
    rad = np.deg2rad(ang)
    for r in [0.18, 0.28]:
        x = center[0] + r * np.cos(rad)
        y = center[1] + r * np.sin(rad)
        ax.plot([center[0], x], [center[1], y], color='#60a5fa', lw=1.2, alpha=0.45)

# Central D(s)=0
ax.text(center[0], center[1], "D(s)=0", ha='center', va='center', fontsize=16, color='cyan', fontweight='bold',
        path_effects=[path_effects.withStroke(linewidth=6, foreground='black')])
ax.text(center[0], center[1]-0.09, "Admissible Manifold S_adm", ha='center', va='center', fontsize=10, color='white')

# Shatter Zone rejection
x_sh = np.linspace(0.57, 0.93, 200)
y_sh = center[1] + 0.10 * np.sin(45 * (x_sh - 0.57)) - 0.06 * (x_sh - 0.57)**2
ax.plot(x_sh, y_sh, color='red', lw=3, linestyle='--')
ax.text(0.80, center[1] + 0.18, "Shatter Zone Eliminated", ha='center', fontsize=11, color='red', fontweight='bold')

# Top tier boxes
ax.add_patch(Rectangle((0.08, 0.84), 0.36, 0.15, fc='#1e40af', ec='#60a5fa', lw=2.2))
ax.text(0.26, 0.915, "V-JEPA\nFeature Extraction\n(egocentric video)", ha='center', va='center', fontsize=9, color='white')

ax.add_patch(Rectangle((0.56, 0.84), 0.36, 0.15, fc='#1e40af', ec='#60a5fa', lw=2.2))
ax.text(0.74, 0.915, "VL-JEPA\nSemantic Grounding", ha='center', va='center', fontsize=9, color='white')

ax.add_patch(Rectangle((0.31, 0.70), 0.38, 0.15, fc='#1e40af', ec='#60a5fa', lw=2.2))
ax.text(0.5, 0.775, "Expert Path\n(C-JEPA causal chain)", ha='center', va='center', fontsize=9, color='white')

# Arrows
ax.add_patch(FancyArrowPatch((0.26, 0.84), (0.74, 0.84), connectionstyle="arc3,rad=0.5", arrowstyle='->', color='#60a5fa', lw=2.5))
ax.add_patch(FancyArrowPatch((0.5, 0.84), (0.5, 0.805), connectionstyle="arc3,rad=0", arrowstyle='->', color='#60a5fa', lw=2.5))
ax.add_patch(FancyArrowPatch((0.5, 0.70), center, connectionstyle="arc3,rad=0", arrowstyle='->', color='#60a5fa', lw=3.2))

# Unitree G1 placeholder
ax.add_patch(Circle((0.5, 0.24), 0.08, color='#60a5fa'))                      # head
ax.add_patch(Rectangle((0.42, 0.11), 0.16, 0.21, fc='#3b82f6', ec='#60a5fa', lw=1.8))  # torso
ax.plot([0.33, 0.42], [0.24, 0.24], color='#60a5fa', lw=8)                   # left arm
ax.plot([0.67, 0.58], [0.24, 0.24], color='#60a5fa', lw=8)                   # right arm
ax.plot([0.45, 0.45], [0.11, -0.02], color='#60a5fa', lw=9)                  # left leg
ax.plot([0.55, 0.55], [0.11, -0.02], color='#60a5fa', lw=9)                  # right leg

# Gauges
ax.add_patch(Arc((0.3, 0.03), 0.32, 0.32, theta1=0, theta2=180, color='#22c55e', lw=7))
ax.text(0.3, -0.03, "+25% Efficiency", ha='center', fontsize=10, color='#22c55e', fontweight='bold')

ax.add_patch(Arc((0.7, 0.03), 0.32, 0.32, theta1=0, theta2=180, color='#ef4444', lw=7))
ax.text(0.7, -0.03, "100% Liability Reduction", ha='center', fontsize=10, color='#ef4444', fontweight='bold')

# Tier titles
ax.text(0.5, 1.00, "H2E Governor: Intent Discovery", ha='center', fontsize=14, fontweight='bold', color='#93c5fd')
ax.text(0.5, 0.64, "Holonomic Guardian: Deterministic Execution", ha='center', fontsize=14, fontweight='bold', color='#93c5fd')
ax.text(0.5, 0.28, "Physical Layer Integration", ha='center', fontsize=14, fontweight='bold', color='#93c5fd')

ax.set_xlim(0, 1)
ax.set_ylim(-0.10, 1.10)
ax.axis('off')

plt.title("Figure 1: H2E-Holonomic Integration Architecture\nTransition via Dispersion Operator D on Unitree G1",
          fontsize=16, color='white', pad=30)

plt.tight_layout()
plt.show()

# To save: uncomment the line below
# plt.savefig('figure1_h2e_holonomic_final.png', dpi=400, bbox_inches='tight', facecolor='#0a0a1f')
```
