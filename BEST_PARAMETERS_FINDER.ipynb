```python
import os
import json
import numpy as np
import pandas as pd
import sqlite3
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import load_model
from ta.trend import MACD
from ta.volatility import BollingerBands, AverageTrueRange
from ta.volume import on_balance_volume
from ta.momentum import RSIIndicator
from collections import deque
import logging
import keras_tuner as kt

from google.colab import drive
drive.mount('/content/gdrive')

# Set up a basic logger for this script
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# --- Helper Functions (as found in your original notebook) ---

def load_sqlite_data(symbol_config):
    try:
        symbol = symbol_config['symbol']
        db_path = symbol_config['db_path']
        table_name = symbol_config['table_name']
        conn = sqlite3.connect(db_path)
        query = f"SELECT * FROM {table_name}"
        df = pd.read_sql_query(query, conn, parse_dates=['timestamp'])
        conn.close()
        if df.empty:
            raise ValueError(f"No data in {db_path}/{table_name}")
        if df['timestamp'].dt.tz is None:
            df['timestamp'] = pd.to_datetime(df['timestamp']).dt.tz_localize('UTC').dt.tz_convert('America/New_York')
        else:
            df['timestamp'] = df['timestamp'].dt.tz_convert('America/New_York')
        df.set_index('timestamp', inplace=True)
        df = df[~df.index.duplicated(keep='last')]
        return df
    except Exception as e:
        logging.error(f"Failed to load data from {db_path}/{table_name}: {e}")
        raise

def calculate_technical_indicators(df):
    df = df.copy()
    df[['open', 'high', 'low', 'close', 'volume']] = df[['open', 'high', 'low', 'close', 'volume']].ffill()
    df['RSI'] = RSIIndicator(df['close'], window=14).rsi()
    macd = MACD(df['close'], window_slow=26, window_fast=12, window_sign=9)
    df['MACD'] = macd.macd()
    df['MACD_Signal'] = macd.macd_signal()
    bb = BollingerBands(df['close'], window=20, window_dev=2)
    df['BB_Upper'] = bb.bollinger_hband()
    df['BB_Lower'] = bb.bollinger_lband()
    df['OBV'] = on_balance_volume(df['close'], df['volume'])
    df['ATR'] = AverageTrueRange(df['high'], df['low'], df['close'], window=14).average_true_range()
    df = df.dropna()
    return df

def prepare_backtest_data(df, look_back):
    df = calculate_technical_indicators(df)
    if len(df) < look_back:
        raise ValueError(f"Insufficient rows after indicators ({len(df)}, need {look_back})")
    return df, df['close'].copy(), df['ATR'].copy()

class PredictionAgent:
    def __init__(self, model_path, look_back=72, default_features=None):
        self.look_back = look_back
        self.default_features = default_features or ['open', 'high', 'low', 'close', 'volume', 'RSI', 'MACD', 'MACD_Signal', 'BB_Upper', 'BB_Lower', 'OBV', 'ATR']
        self.scaler = MinMaxScaler(feature_range=(0, 1))
        self.model = load_model(model_path)
        self.expected_features = self.model.input_shape[-1]
        self.features = self.default_features[:min(len(self.default_features), self.expected_features)]
        self.predict_fn = tf.function(self.model, input_signature=[tf.TensorSpec(shape=(None, self.look_back, self.expected_features), dtype=tf.float32)])
    def preprocess(self, df):
        data = df[self.features].to_numpy(dtype=np.float32)
        scaled = self.scaler.fit_transform(data)
        if len(self.features) < self.expected_features:
            padding = np.zeros((scaled.shape[0], self.expected_features - len(self.features)), dtype=np.float32)
            scaled = np.hstack([scaled, padding])
        X = scaled[-self.look_back:].reshape(1, self.look_back, self.expected_features)
        X = tf.convert_to_tensor(X, dtype=tf.float32)
        return X
    def execute(self, df):
        X = self.preprocess(df)
        pred_probs = self.predict_fn(X)
        return np.asarray(pred_probs, dtype=np.float32)

def run_backtest_for_tuner(trade_params, prediction_agent, df_test, original_prices_test, original_atr_test, symbol="Generic"):
    initial_capital = 2500.0
    capital = initial_capital
    qty = 0.0
    num_trades = 0
    num_wins = 0
    num_losses = 0
    in_position = False
    entry_price = 0.0
    entry_t = 0
    last_exit_t = -6
    recent_pl = deque(maxlen=5)

    look_back = prediction_agent.look_back

    # Volatility filter: Use rolling 25th percentile ATR
    atr_threshold = original_atr_test.rolling(window=1000, min_periods=1).quantile(0.25)
    # Trend filter: 200-period SMA
    sma200 = original_prices_test.rolling(window=200, min_periods=1).mean()

    for t in range(look_back, len(df_test)):
        current_price = original_prices_test.iloc[t]
        atr = original_atr_test.iloc[t]

        # --- Adjusted Filtering Logic ---
        # Get the latest data window
        window_df = df_test.iloc[max(0, t - look_back):t]
        if len(window_df) < look_back:
            continue

        try:
            pred_probs = prediction_agent.execute(window_df)
            pred_class = np.argmax(pred_probs, axis=1)[0]
            confidence = pred_probs[0, pred_class]
        except Exception as e:
            continue

        # Relaxed Filters for LDO and TAO to increase trade frequency
        is_atr_valid = True
        if symbol in ['LDO/USD', 'TAO/USD']:
            # Relax the ATR filter
            if atr < atr_threshold.iloc[t] * 0.5:
                is_atr_valid = False
        else:
            if atr < atr_threshold.iloc[t]:
                is_atr_valid = False

        if not is_atr_valid:
            continue


        # Main Entry/Exit Logic
        if not in_position and pred_class == 1 and confidence >= trade_params['CONFIDENCE_THRESHOLD']:
            # Use MAX_POSITION_SIZE from tuned parameters
            position_size_usd = capital * trade_params['MAX_POSITION_SIZE']
            qty = position_size_usd / current_price

            # Check for a minimum trade size
            min_volume = {'ETH/USD': 0.002, 'SOL/USD': 0.1, 'LDO/USD': 10, 'TAO/USD': 0.01}.get(symbol, 0.002)
            if qty < min_volume:
                continue

            capital -= position_size_usd
            entry_price = current_price
            entry_t = t
            in_position = True
            num_trades += 1

        elif in_position:
            exit_reason = None
            exit_price = current_price

            # Take Profit
            if current_price >= entry_price + atr * trade_params['ATR_MULTIPLIER_TP']:
                exit_reason = 'TP'
                exit_price = entry_price + atr * trade_params['ATR_MULTIPLIER_TP']
            # Stop Loss
            elif current_price <= entry_price - atr * trade_params['ATR_MULTIPLIER_SL']:
                exit_reason = 'SL'
                # Fix: Use trade_params['ATR_MULTIPLIER_SL'] instead of trade_multiplier_sl
                exit_price = entry_price - atr * trade_params['ATR_MULTIPLIER_SL']
            # Sell Signal
            elif pred_class == 2 and confidence >= trade_params['CONFIDENCE_THRESHOLD']:
                exit_reason = 'Sell_Signal'

            if exit_reason:
                profit_or_loss = (exit_price - entry_price) * qty
                capital += (entry_price * qty) + profit_or_loss
                recent_pl.append(profit_or_loss)

                if profit_or_loss > 0:
                    num_wins += 1
                else:
                    num_losses += 1

                in_position = False
                qty = 0.0
                last_exit_t = t

    # Close open position at end of backtest
    if in_position:
        profit_or_loss = (original_prices_test.iloc[-1] - entry_price) * qty
        capital += (original_prices_test.iloc[-1] * qty)
        if profit_or_loss > 0:
            num_wins += 1
        else:
            num_losses += 1

    total_return = ((capital - initial_capital) / initial_capital) * 100 if initial_capital > 0 else 0
    # Return a metric that Keras Tuner can optimize (e.g., total return)
    return total_return, num_trades

def build_trading_strategy(hp, prediction_agent, df_test, original_prices_test, original_atr_test, symbol):
    # Define hyperparameters to tune
    confidence_threshold = hp.Float('confidence_threshold', min_value=0.3, max_value=0.9, step=0.05)
    atr_multiplier_tp = hp.Float('atr_multiplier_tp', min_value=1.0, max_value=5.0, step=0.5)
    atr_multiplier_sl = hp.Float('atr_multiplier_sl', min_value=0.5, max_value=3.0, step=0.5)
    max_position_size = hp.Float('max_position_size', min_value=0.01, max_value=0.1, step=0.01)


    # Create a dictionary of trading parameters
    trade_params = {
        'CONFIDENCE_THRESHOLD': confidence_threshold,
        'ATR_MULTIPLIER_TP': atr_multiplier_tp,
        'ATR_MULTIPLIER_SL': atr_multiplier_sl,
        'MAX_POSITION_SIZE': max_position_size
    }

    # Run the backtest with the current parameters
    total_return, num_trades = run_backtest_for_tuner(
        trade_params, prediction_agent, df_test, original_prices_test, original_atr_test, symbol
    )

    # Keras Tuner optimizes based on a single scalar metric.
    # We'll use total return as the objective.
    # We can also add a penalty for too few trades if needed, but let's start simple.
    objective_metric = total_return if num_trades > 5 else -1000 # Penalize strategies with very few trades

    # Return a dummy Keras model with the objective metric as a custom metric.
    # This is a workaround to use Keras Tuner for parameter optimization of a non-Keras function.
    # The "model" itself doesn't do any prediction; its "training" is the backtest.
    model = tf.keras.Sequential()
    model.add(tf.keras.layers.Input(shape=(1,))) # Dummy input layer
    model.add(tf.keras.layers.Lambda(lambda x: objective_metric)) # Use Lambda layer to output the metric

    model.compile(loss='mse') # Compile with a dummy loss and no metrics

    return model


# --- Main Execution ---
if __name__ == "__main__":
    # Define a single symbol to optimize
    symbol_config = {
        "symbol": "LDO/USD",
        "model_path": '/content/gdrive/MyDrive/TradingBotLogs/crypto_model_retrained_500epochs_v2_LDO.keras',
        "params": {"CONFIDENCE_THRESHOLD": 0.4, "ATR_MULTIPLIER_TP": 2.0, "ATR_MULTIPLIER_SL": 1.0, "MAX_POSITION_SIZE": 0.03},
        "db_path": '/content/gdrive/MyDrive/TradingBotLogs/ohlcv_data_LDO.db',
        "table_name": 'ldousd_1h_data'
    }

    print(f"\n--- Starting Parameter Optimization for {symbol_config['symbol']} using Keras Tuner ---")

    try:
        df = load_sqlite_data(symbol_config)
        df_test, original_prices_test, original_atr_test = prepare_backtest_data(df, 72)
        prediction_agent = PredictionAgent(symbol_config['model_path'])
    except Exception as e:
        print(f"Error preparing data for optimization: {e}")
        exit()

    # Create a Keras Tuner Hyperband tuner
    tuner = kt.Hyperband(
        lambda hp: build_trading_strategy(hp, prediction_agent, df_test, original_prices_test, original_atr_test, symbol_config['symbol']),
        objective=kt.Objective('total_return', direction='max'),
        max_epochs=10,  # Maximum number of epochs for Hyperband
        factor=3,  # Factor by which the number of epochs is reduced and the number of models is increased
        directory='/content/gdrive/MyDrive/TradingBotLogs/keras_tuner_dir',  # Directory to save tuning results
        project_name=f'trading_param_tuning_{symbol_config["symbol"].replace("/", "_")}'
    )

    # Run the tuner. Since we are not training a Keras model in the traditional sense,
    # we pass dummy data to the tuner's search method. The actual "training" happens
    # within the build_trading_strategy function when run_backtest_for_tuner is called.
    # We need at least one sample for the tuner to run.
    dummy_data = np.zeros((len(df_test) - prediction_agent.look_back, 1)) # Dummy input data
    dummy_labels = np.zeros((len(df_test) - prediction_agent.look_back,)) # Dummy labels

    # Use the entire dataset for tuning since the backtest uses the whole series
    tuner.search(dummy_data, dummy_labels, epochs=1, batch_size=1)


    # Get the best hyperparameters
    best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]

    print(f"\nOptimization Complete using Keras Tuner.")
    print(f"Best Parameters for {symbol_config['symbol']}:")
    print(f"  CONFIDENCE_THRESHOLD: {best_hps.get('confidence_threshold'):.2f}")
    print(f"  ATR_MULTIPLIER_TP: {best_hps.get('atr_multiplier_tp'):.2f}")
    print(f"  ATR_MULTIPLIER_SL: {best_hps.get('atr_multiplier_sl'):.2f}")
    print(f"  MAX_POSITION_SIZE: {best_hps.get('max_position_size'):.2f}")


    # You can optionally run the backtest one more time with the best parameters to get the final metrics
    best_trade_params = {
        'CONFIDENCE_THRESHOLD': best_hps.get('confidence_threshold'),
        'ATR_MULTIPLIER_TP': best_hps.get('atr_multiplier_tp'),
        'ATR_MULTIPLIER_SL': best_hps.get('atr_multiplier_sl'),
        'MAX_POSITION_SIZE': best_hps.get('max_position_size')
    }
    final_return, final_trades = run_backtest_for_tuner(
        best_trade_params, prediction_agent, df_test, original_prices_test, original_atr_test, symbol_config['symbol']
    )
    print(f"Final Total Return with Best Parameters: {final_return:.2f}%")
    print(f"Number of Trades with Best Parameters: {final_trades}")
```
