```python
import matplotlib.pyplot as plt
import numpy as np
import random

# Frequency bands in original order (top to bottom)
frequency_bands = ["High Gamma", "Gamma", "Beta", "Alpha", "Theta", "Delta"]

# Simulated heritability data for MZ and DZ twins (6 bands x 10 regions each)
heritability_data_mz = np.array([[random.uniform(0, 1) for _ in range(10)] for _ in range(len(frequency_bands))])  # MZ Twins
heritability_data_dz = np.array([[random.uniform(0, 0.5) for _ in range(10)] for _ in range(len(frequency_bands))])  # DZ Twins (lower range to reflect lower similarity)

# Create a figure with two subplots side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Heritability heatmap for MZ twins
im1 = ax1.imshow(heritability_data_mz, cmap='hot', aspect='auto')
ax1.set_yticks(np.arange(len(frequency_bands)))
ax1.set_yticklabels(frequency_bands[::-1])  # High to low order
ax1.set_xticks([])
ax1.set_title("Heritability - MZ Twins")

# Heritability heatmap for DZ twins
im2 = ax2.imshow(heritability_data_dz, cmap='hot', aspect='auto')
ax2.set_yticks(np.arange(len(frequency_bands)))
ax2.set_yticklabels(frequency_bands[::-1])  # High to low order
ax2.set_xticks([])
ax2.set_title("Heritability - DZ Twins")

# Add colorbars
plt.colorbar(im1, ax=ax1, label="Heritability (0-1)")
plt.colorbar(im2, ax=ax2, label="Heritability (0-1)")

plt.tight_layout()
plt.show()
```
