```python
import os
from google.colab import userdata
from crewai import Agent, Task, Crew
from langchain_google_genai import ChatGoogleGenerativeAI

# 1. Securely fetch the key from Colab Secrets
try:
    api_key = userdata.get('GEMINI')
    # Use standard 2026 environment variable naming
    os.environ["GOOGLE_API_KEY"] = api_key
    print("‚úÖ Secret 'GEMINI' loaded successfully.")
except Exception as e:
    print("‚ùå ERROR: Secret 'GEMINI' not found in Colab (üîë).")
    raise e

# 2. Update to the 2026 Stable Model: gemini-2.5-flash
# Note: gemini-1.5-flash was retired in late 2025.
custom_llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    google_api_key=api_key,
    temperature=0.3
)

# 3. Define the Analyst Agent
analyst = Agent(
    role='AGI Capability Researcher',
    goal='Contrast the 2026 Turing-AGI Test with traditional Turing tests.',
    backstory='You are an expert in AI labor economics, as seen in The Batch Jan 2026 issue.',
    llm=custom_llm,
    verbose=False,
    allow_delegation=False
)

# 4. Define the Utility-Based Task
task = Task(
    description='Analyze why Andrew Ng‚Äôs new test focuses on "economically useful work" over "imitation."',
    expected_output='A summary table comparing the tests on Deception vs. Utility.',
    agent=analyst,
    human_input=False
)

# 5. The Final Crew
crew = Crew(
    agents=[analyst],
    tasks=[task],
    share_crew=False # Mandatory: Prevents the Thread-9 AttributeError
)

print("\n### KICKING OFF 2026 AGENTIC WORKFLOW ###\n")
result = crew.kickoff()

print("\n\n##############################")
print("## FINAL AUDITED REPORT ##")
print("##############################\n")
print(result)
```
