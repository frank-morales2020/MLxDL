```python
import openai
import os
from dotenv import load_dotenv
load_dotenv()
import colab_env

from openai import OpenAI
client = OpenAI(api_key = os.getenv("OPENAI_API_KEY"))
#client = OpenAI()

# Set your OpenAI API key
openai.api_key =  os.getenv("OPENAI_API_KEY")

# Define the cache with information about planets
cache = {
    "What is the closest planet to the Sun?": "Mercury is the closest planet to the Sun.",
    "What is the largest planet in our solar system?": "Jupiter is the largest planet in our solar system.",
    "How many moons does Mars have?": "Mars has two moons, Phobos and Deimos.",
    "What is the Great Red Spot?": "The Great Red Spot is a giant storm on Jupiter that has been raging for centuries."
}

# Function to generate text with cache augmentation

def generate_with_cache(prompt, max_length=500):
    # Search for relevant cache entries in the prompt (case-insensitive)
    relevant_cache = ""
    for key, value in cache.items():
        if key.lower() in prompt.lower():
            relevant_cache += f"{key}: {value}\n"

    # Prepend the relevant cache entries to the prompt
    if relevant_cache:
        augmented_prompt = relevant_cache + prompt
    else:
        augmented_prompt = prompt

    # Generate text using the GPT-3.5-turbo model using the new API structure
    response = client.chat.completions.create( # Use client.chat.completions.create
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": augmented_prompt}
        ],
        max_tokens=max_length,
        temperature=0.7  # Add temperature for more diverse responses

    )

    # Extract and return the generated text
    generated_text = response.choices[0].message.content # Access content using response.choices[0].message.content
    return generated_text



# Define the AI agent class
class AIAgent:
    def __init__(self, cache):
        self.cache = cache
        self.history = []

    def respond(self, user_input):
        # Add user input to the history
        self.history.append({"role": "user", "content": user_input})
        self.cache["history"] = self.history  # Store history in the cache

        # Generate response with cache augmentation
        response = generate_with_cache(user_input)

        # Add the agent's response to the history
        self.history.append(response)

        return response

    def display_history(self):
        print('\n')
        print("\nConversation History:")
        print('\n')
        for i, utterance in enumerate(self.history):
            print(f"{'User' if i % 2 == 0 else 'Agent'}: {utterance}")


# Create an AI agent instance
agent = AIAgent(cache)

# Interact with the agent
while True:
    print('\n')
    user_input = input("User: ")
    if user_input.lower() == "exit":
        break
    response = agent.respond(user_input)
    print(f"Agent: {response}")


# Display the conversation history when the interaction ends
agent.display_history()

# What is the Great Red Spot?
# Tell me about the moons of Mars.

# Can you write an essay about the Great Red Spot and the moons of Mars, based on the information we just discussed?"
```
