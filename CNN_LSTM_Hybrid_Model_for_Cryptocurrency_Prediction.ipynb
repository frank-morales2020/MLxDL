```
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, LSTM, Dense, Flatten, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import matplotlib.pyplot as plt
import sqlite3 # Import for SQLite database interaction

# --- 1. Data Loading (USING YOUR ETH SQLITE DATABASE) ---
# Define your SQLite database path and table name for ETH
SQLITE_DB_PATH = '/content/gdrive/MyDrive/TradingBotLogs/ohlcv_data.db'
SQLITE_TABLE_NAME = 'ethusd_1h_data'

print(f"--- Attempting to load data from SQLite: {SQLITE_DB_PATH} - Table: {SQLITE_TABLE_NAME} ---")

try:
    conn = sqlite3.connect(SQLITE_DB_PATH)
    # Fetch data, assuming columns like 'timestamp', 'open', 'high', 'low', 'close', 'volume'
    # Adjust column names if they are different in your table
    query = f"SELECT timestamp, open, high, low, close, volume FROM {SQLITE_TABLE_NAME} ORDER BY timestamp ASC"
    df = pd.read_sql_query(query, conn, parse_dates=['timestamp'], index_col='timestamp')
    conn.close()

    # Rename 'close' to 'ETH_Close' for consistency with the example target
    df = df.rename(columns={'close': 'ETH_Close'})

    # If you have data for BTC, SOL, TAO in the *same* database or separate, you would load them here
    # and then merge them into this DataFrame 'df' on the 'timestamp' index.
    # For now, we're focusing on ETH data as per your request.
    # Placeholder for other cryptos if they were to be loaded:
    # df['BTC_Close'] = ...
    # df['SOL_Close'] = ...
    # df['TAO_Close'] = ...

    print("--- Successfully loaded ETH data from SQLite ---")
    print("--- ETH Data Sample ---")
    print(df.head())
    print("\n--- ETH Data Description ---")
    print(df.describe())
    print("-" * 40)

except Exception as e:
    print(f"Error loading data from SQLite: {e}")
    print("Falling back to synthetic data for demonstration.")
    # --- Fallback: Data Generation (if SQLite loading fails or for other cryptos not loaded) ---
    num_days = 10 * 365 # Approx. 10 years of daily data (adjusted if using 1h data)
    dates = pd.date_range(start='2015-01-01', periods=num_days, freq='D')

    np.random.seed(42) # for reproducibility

    base_prices = {'BTC': 500, 'ETH': 50, 'SOL': 10, 'TAO': 5}
    volatilities = {'BTC': 0.03, 'ETH': 0.05, 'SOL': 0.07, 'TAO': 0.10}

    data = {}
    for crypto in ['BTC', 'ETH', 'SOL', 'TAO']:
        prices = [base_prices[crypto]]
        for _ in range(1, num_days):
            if crypto == 'BTC': drift = 0.0001
            elif crypto == 'ETH': drift = 0.0003
            elif crypto == 'SOL': drift = 0.0005
            elif crypto == 'TAO': drift = 0.0007
            change = np.random.normal(drift, volatilities[crypto])
            new_price = prices[-1] * (1 + change)
            prices.append(max(1.0, new_price))
        data[f'{crypto}_Close'] = prices

    df = pd.DataFrame(data, index=dates)
    print("--- Synthetic Data Sample (fallback) ---")
    print(df.head())
    print("\n--- Synthetic Data Description (fallback) ---")
    print(df.describe())
    print("-" * 40)


# --- 2. Data Preprocessing ---
# We'll predict the 'ETH_Close' price as an example, using its OHLCV data as features.
target_column = 'ETH_Close'
# Ensure 'target_column' is in df.columns
if target_column not in df.columns:
    print(f"Error: Target column '{target_column}' not found in DataFrame. Please check your database schema or data loading logic.")
    # If the target column isn't found, we can't proceed. Exit or use a synthetic fallback.
    exit() # Or handle this more gracefully, e.g., by using a different target

# Features will include 'open', 'high', 'low', 'ETH_Close', 'volume' for ETH
features = ['open', 'high', 'low', 'ETH_Close', 'volume']
# Filter out any features that might not exist in the loaded data (e.g. if using fallback data)
features = [f for f in features if f in df.columns]

# Ensure the DataFrame is sorted by timestamp
df = df.sort_index()

# Scale the data to be between 0 and 1, which is crucial for neural networks
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(df[features])

# Create sequences for the CNN-LSTM
# A sequence will consist of 'look_back' time steps, and we'll predict the next step.
look_back = 60 # Number of previous time steps (e.g., hours if using 1h data) to consider for prediction
prediction_horizon = 1 # Number of future steps to predict (e.g., predict the next hour's close)

X, y = [], []
for i in range(len(scaled_data) - look_back - prediction_horizon + 1):
    X.append(scaled_data[i:(i + look_back), :])
    y.append(scaled_data[i + look_back + prediction_horizon - 1, features.index(target_column)]) # Predict the target column

X = np.array(X)
y = np.array(y)

# Split data into training and testing sets
# It's important to split time series data sequentially, not randomly.
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

print(f"\nShape of X_train: {X_train.shape}")
print(f"Shape of y_train: {y_train.shape}")
print(f"Shape of X_test: {X_test.shape}")
print(f"Shape of y_test: {y_test.shape}")
print("-" * 40)

# --- 3. Build the CNN-LSTM Hybrid Model ---
# The input shape for the model is (timesteps, features)
n_features = X_train.shape[2] # Number of features (open, high, low, close, volume for ETH)
n_timesteps = X_train.shape[1] # look_back value

model = Sequential([
    # CNN part for feature extraction
    Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(n_timesteps, n_features)),
    MaxPooling1D(pool_size=2),
    Dropout(0.2), # Adding dropout for regularization

    Conv1D(filters=128, kernel_size=3, activation='relu'),
    MaxPooling1D(pool_size=2),
    Dropout(0.2),

    # The output of MaxPooling1D is already 3D (batch, steps, features),
    # which is the expected input for an LSTM layer.
    LSTM(100, activation='relu', return_sequences=False), # False because it's the last LSTM before dense
    Dropout(0.3),

    Dense(50, activation='relu'),
    Dense(1) # Output a single value (the predicted price)
])

model.compile(optimizer='adam', loss='mse') # Mean Squared Error is common for regression tasks

print("\n--- Model Summary ---")
model.summary()
print("-" * 40)

# --- 4. Model Training ---
# Define callbacks for better training
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=0.0001)

history = model.fit(X_train, y_train,
                    epochs=100, # You might need more epochs for real data
                    batch_size=32,
                    validation_split=0.1, # Use a portion of training data for validation
                    callbacks=[early_stopping, reduce_lr],
                    verbose=1)

# Plot training history
plt.figure(figsize=(12, 6))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss (MSE)')
plt.legend()
plt.grid(True)
plt.show()

# --- 5. Model Evaluation ---
# Evaluate the model on the test set
loss = model.evaluate(X_test, y_test, verbose=0)
print(f"\nTest Loss (MSE): {loss:.6f}")

# Make predictions
y_pred_scaled = model.predict(X_test)

# Inverse transform the predictions and actual values to get them back to original scale
# To inverse transform, we need to create a dummy array with the same number of features
# as the scaled_data, and then replace the target column with our predictions/actuals.
dummy_array = np.zeros((len(y_pred_scaled), n_features))
dummy_array[:, features.index(target_column)] = y_pred_scaled.flatten()
y_pred = scaler.inverse_transform(dummy_array)[:, features.index(target_column)]

dummy_array_actual = np.zeros((len(y_test), n_features))
dummy_array_actual[:, features.index(target_column)] = y_test.flatten()
y_actual = scaler.inverse_transform(dummy_array_actual)[:, features.index(target_column)]

# Plot actual vs. predicted prices
plt.figure(figsize=(14, 7))
plt.plot(y_actual, label='Actual Price', color='blue', alpha=0.7)
plt.plot(y_pred, label='Predicted Price', color='red', linestyle='--', alpha=0.7)
plt.title(f'Actual vs. Predicted {target_column} Price')
plt.xlabel('Time Step (Test Data)')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()

# --- Example of making a single prediction for the next day/hour ---
# Take the last 'look_back' time steps from the original scaled data
last_look_back_data = scaled_data[-look_back:]
last_look_back_data = last_look_back_data.reshape(1, look_back, n_features) # Reshape for model input

next_prediction_scaled = model.predict(last_look_back_data)

# Inverse transform the single prediction
dummy_array_single = np.zeros((1, n_features))
dummy_array_single[:, features.index(target_column)] = next_prediction_scaled.flatten()
next_prediction = scaler.inverse_transform(dummy_array_single)[:, features.index(target_column)]

print(f"\nPredicted {target_column} price for the next time step: {next_prediction[0]:.2f}")
```
