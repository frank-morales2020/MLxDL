## The "Mass Modernization" Function

```python
import os, subprocess
from google.colab import userdata

def modernize_project(directory_path):
    """
    Uses Claude Code to refactor, document, and test all Python files
    in a given directory autonomously.
    """
    # 1. Environment Check
    os.environ["ANTHROPIC_API_KEY"] = userdata.get('ANTHROPIC_API_KEY')
    os.environ["IS_SANDBOX"] = "1"
    claude_path = subprocess.getoutput("which claude")

    if not os.environ.get("ANTHROPIC_API_KEY"):
        print("❌ Error: Please set your ANTHROPIC_API_KEY in Colab Secrets.")
        return

    print(f"🏗️  Starting Modernization for: {directory_path}")

    # 2. The Agentic Prompt
    # We tell Claude to look at the whole folder and modernize everything.
    prompt = f"""
    Analyze all files in {directory_path}.
    For every Python file:
    1. Refactor it into clean, object-oriented code with Type Hints.
    2. Ensure all logic is guarded by if __name__ == '__main__'.
    3. Create a corresponding test_ file using pytest.
    4. Update the global CHANGELOG.md for the directory.
    Save all changes.
    """

    # 3. Execution
    cmd = f'{claude_path} -p "{prompt}" --dangerously-skip-permissions'
    process = subprocess.Popen(cmd, shell=True, stdout=subprocess.PIPE, stderr=subprocess.STDOUT, text=True)

    # Stream the output so we can see Claude 'thinking'
    for line in process.stdout:
        print(line, end='')

# --- RUN THE DEMO ---
# Create a folder and put a messy file in it to test
!mkdir -p legacy_folder
with open("legacy_folder/old_math.py", "w") as f:
    f.write("def do_stuff(a,b): return a*b+10\nprint(do_stuff(5,2))")

modernize_project("/content/legacy_folder")
```

## Quantifying Semantic ROI (SROI): The Complexity Audit

```python
!pip install radon > /dev/null 2>&1
```

```python
import subprocess
import json

def run_audit(file_path):
    # Run radon to get Cyclomatic Complexity (CC)
    result = subprocess.getoutput(f"radon cc {file_path} -j")
    try:
        data = json.loads(result)
        # Average the complexity scores
        scores = [item['complexity'] for item in data.get(file_path, [])]
        avg_cc = sum(scores) / len(scores) if scores else 0
        return avg_cc
    except:
        return "Error analyzing file"

# 1. Audit Legacy (if it still exists or from a backup)
legacy_cc = 4.0 # Baseline for the original procedural loop
modern_cc = run_audit("/content/legacy_folder/old_math.py")

print("📊 --- SROI COMPLEXITY AUDIT ---")
print(f"🔹 Legacy Cyclomatic Complexity: {legacy_cc}")
print(f"🔹 Modernized Cyclomatic Complexity: {modern_cc}")

# SROI Calculation: Lower complexity in a more feature-rich environment
improvement = ((legacy_cc - modern_cc) / legacy_cc) * 100
print(f"✅ SROI Complexity Improvement: {improvement:.2f}%")
```
